### Hagan girar una imagen


Primero se calculan las dimensiones del : ancho , alto  y el cento dela figura a rotar, luego se establece una matriz de rotaci√≥n con el angulo a rotar, para luego usar una transformaci√≥n con warpAffine, con los valores del ancho y la altura.

```python
(altura, ancho)=img2.shape[:2] 
(cX, cY) = (ancho//2, altura//2) 

M = cv2.getRotationMatrix2D((cX, cY), 65, 1.0) 
rotacion = cv2.warpAffine(img2, M, (ancho, altura)) 
plt.imshow(rotacion)
plt.show()
```

### Dibujen un c√≠rculo sobre las caras de dicha figura y agreguen un texto que describa que es (persona, perro, gato).


Circulo:
Sintaxis: 
imgWC = cv2.circle(img_G,(850,110),100,(0,0,255),15,cv2.LINE_AA)

- El primer argumento de la funci√≥n circle() es la imagen que queremos dibujar del c√≠rculo.

- El segundo argumento es el punto central del c√≠rculo, y su valor debe ser un punto que contenga los valores de los ejes x e y de un p√≠xel que queremos usar como centro del c√≠rculo.

- El tercer argumento es el radio del c√≠rculo y su valor puede ser cualquier n√∫mero entero.
El cuarto argumento es el color del c√≠rculo, y debe ser un valor de triplete BGR como (0,0,255) para el color rojo.

Letra:
Sintaxis: 

 

cv2.putText (imagen, texto, organizaci√≥n, fuente, escala de fuente, color [, grosor [, tipo de l√≠nea [, origen inferior izquierdo]]])
Texto: cadena de texto que se va a dibujar.

- Texto: cadena de texto que se va a dibujar

- Organizaci√≥n: son las coordenadas de la esquina inferior izquierda de la cadena de texto en la imagen. Las coordenadas se representan como tuplas de dos valores, es decir ( valor de la coordenada X , valor de la coordenada Y ).


- Fuente: Denota el tipo de fuente. Algunos tipos de fuentes son FONT_HERSHEY_SIMPLEX, FONT_HERSHEY_PLAIN, etc.


- FontScale: factor de escala de fuente que se multiplica por el tama√±o base espec√≠fico de la fuente.


- Color: es el color de la cadena de texto que se va a dibujar. Para BGR , pasamos una tupla. por ejemplo: (0,255,255) para el color amarillo.


- Grosor: Es el grosor de la l√≠nea en px .

```python
img_G = cv2.imread('img.jpeg')

img_G = cv2.cvtColor(img_G, cv2.COLOR_BGR2RGB)
#persona

imgWC = cv2.circle(img_G,(850,110),100,(0,0,255),15,cv2.LINE_AA)
cv2.putText(img_G,"Persona",(800,110),cv2.FONT_HERSHEY_COMPLEX,2,(0,0,255),3)
#gato

imgWC = cv2.circle(img_G,(500,380),80,(0,255,0),15,cv2.LINE_AA)

cv2.putText(img_G,"Gato",(450,380),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,0),3)
#perro

imgWC = cv2.circle(img_G,(750,300),130,(255,0,0),15,cv2.LINE_AA)

cv2.putText(img_G,"Perro",(700,300),cv2.FONT_HERSHEY_COMPLEX,2,(255,0,0),3)

plt.imshow(imgWC)
```

### En una imagen distinta al anterior creen una pent√°gono sobre las caras de las figuras y agreguen un texto que describa que es (persona, perro, gato).


```python
img_G2 = cv2.imread('img_G2.jpg')
img_G2 = cv2.cvtColor(img_G2, cv2.COLOR_BGR2RGB)

def drawPentagono(img, x, y, tam, color) -> None:
    cv2.line(img, (x, y-tam), (x-tam, y-20), color, 5)
    cv2.line(img, (x-tam, y-20), (x-tam+20, y+tam), color, 5)
    cv2.line(img, (x-tam+20, y+tam), (x+tam-20, y+tam), color, 5)
    cv2.line(img, (x+tam-20, y+tam), (x+tam, y-20), color, 5)
    cv2.line(img, (x+tam, y-20), (x, y-tam), color, 5)
    
#persona
drawPentagono(img_G2, 400, 150, 130, (0,0,255))
cv2.putText(img_G2,"Persona",(300, 150),cv2.FONT_HERSHEY_COMPLEX,2,(0,0,255),3)


#gato

drawPentagono(img_G2, 230, 250, 80, (0,255,0))
cv2.putText(img_G2,"Gato",(150, 250),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,0),3)

#perro
drawPentagono(img_G2, 600, 200, 80, (255,0,0))
cv2.putText(img_G2,"Perro",(550, 200),cv2.FONT_HERSHEY_COMPLEX,2,(255,0,0),3)
plt.imshow(img_G2)
```

### Creen un programa que permita dibujar diferentes tipos de figuras geom√©tricas con el uso de eventos de mouse para poder hacer dibujos y luego guarden la imagen.


```python
import numpy as np
import cv2


def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),100,(255,0,0),-1)
        

def draw_rectangle(event,x,y,flags,param):
    global ix,iy
    if event == cv2.EVENT_LBUTTONDOWN:
        ix,iy=x,y
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.rectangle(img,(ix,iy),(x,y),(255,0,0),-1)

def draw_line(event,x,y,flags,param):
    global ix,iy
    if event == cv2.EVENT_LBUTTONDOWN:
        ix,iy=x,y
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.line(img,(ix,iy),(x,y),(255,0,0),5)

        
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

while(1):
    cv2.imshow('image',img)
    k = cv2.waitKey(20) & 0xFF
    if k == 27:
        break
    elif k == ord('c'):
        cv2.setMouseCallback('image',draw_circle)
    elif k == ord('r'):
        cv2.setMouseCallback('image',draw_rectangle)
    elif k == ord('l'):
        cv2.setMouseCallback('image',draw_line)
    elif k == ord('d'):
        cv2.setMouseCallback('image',dibujo_draw_circle)

        
cv2.destroyAllWindows()

```
