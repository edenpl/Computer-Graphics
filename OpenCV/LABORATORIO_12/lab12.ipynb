### Creen un arreglo de 5x5 donde los números sean aleatorios del 1 al 10


Se crea una matriz con dos arreglos de cinco espacios, para luego llenar con el valor aleatorio e imprimirlos de forma matricial.

```python
matrix = []
for i in range(5):
  matrix.append([])
  for j in range(5):
    matrix[i].append(random.randint(0,10))

for i in range(5):
        for j in range(5):
            print(matrix[i][j], end=" ")
        print()
```

### Cambien la imagen a escala de grises


La escala de grises es el proceso de convertir una imagen de otros espacios de color como RGB a tonos de gris. Varía entre el negro completo y el blanco completo.

Se utiliza COLOR_BGR2GRAY.

```python
img_gray1 = cv2.imread("D:\py\lab\lab12\img\img.jpg")
img_gray1 = cv2.cvtColor(img_gray1, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray1, cmap='gray')
```

Código más rápido:

```python
img_gray2 = cv2.imread("D:\py\lab\lab12\img\img.jpg",0)
plt.imshow(img_gray2, cmap='gray')
```
